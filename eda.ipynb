```python
df.columns = df.columns.str.strip()
```

```python
disturbance_type_counts = df["disturbance_type"].value_counts()
disturbance_type_counts
```

1 &rarr; Undisturbed

2 &rarr; Naturally Disturbed (animals, etc.)

3 &rarr; Anthropogenically Disturbed (strictly humans)

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.barplot(ax=ax, data=disturbance_type_counts, color="salmon")

ax.set_title("Distribution of Disturbance Types", fontsize=18, fontweight="bold")
ax.set_xlabel("Disturbance Type", fontweight="bold", fontsize=14)
ax.set_ylabel("")

ax.set_xticks([0, 1, 2])
ax.set_xticklabels(
    ["Undisturbed", "Naturally Disturbed", "Anthropogenically Disturbed"]
)

fig.tight_layout()
```

```python
stability_disturbance = df.groupby("surface_stability")[
    "disturbance_type"
].value_counts()
stability_disturbance
```

```python
stability_disturbance = stability_disturbance.reset_index(name="count")
stability_disturbance
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.scatterplot(
    data=stability_disturbance,
    ax=ax,
    y="surface_stability",
    x="disturbance_type",
    size="count",
    hue="count",
    palette="flare",
    sizes=(100, 1000),
    edgecolor="black",
    alpha=0.8,
)

ax.set_title(
    "Frequency of Combinations: Surface Stability vs Disturbance",
    fontsize=14,
    pad=15,
    fontweight="bold",
)
ax.set_xlabel("Disturbance Type", fontsize=12)
ax.set_ylabel("Surface Stability", fontsize=12)

ax.set_xticks([1, 2, 3])
ax.set_xticklabels(
    ["Undisturbed", "Naturally Disturbed", "Anthropogenically Disturbed"]
)

ax.grid(alpha=0.3, linestyle="--")

ax.legend(bbox_to_anchor=(1.05, 1), loc="upper left", title="Count")

fig.tight_layout()
```
